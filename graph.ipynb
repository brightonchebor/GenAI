```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
A = 1          # Amplitude of the waves
k = 2 * np.pi  # Wave number (2*pi for one wavelength in the range)
omega = 2 * np.pi  # Angular frequency (2*pi for one cycle in the range)
phi = np.pi / 2  # Phase difference (in radians)

# Create an array of x values
x = np.linspace(0, 2, 1000)  # From 0 to 2 meters

# Calculate the two waves
y1 = A * np.sin(k * x - omega * 0 + 0)          # Wave 1
y2 = A * np.sin(k * x - omega * 0 + phi)       # Wave 2

# Calculate the resultant wave using superposition
resultant_wave = y1 + y2

# Plotting
plt.figure(figsize=(12, 6))

# Plot Wave 1
plt.plot(x, y1, label='Wave 1: $A \sin(kx - \omega t)$', color='blue')

# Plot Wave 2
plt.plot(x, y2, label='Wave 2: $A \sin(kx - \omega t + \phi)$', color='orange')

# Plot Resultant Wave
plt.plot(x, resultant_wave, label='Resultant Wave', color='green', linewidth=2)

# Adding titles and labels
plt.title('Superposition of Two Waves with Phase Difference')
plt.xlabel('Position (x)')
plt.ylabel('Amplitude')
plt.axhline(0, color='black', lw=0.5, ls='--')  # X-axis
plt.axvline(0, color='black', lw=0.5, ls='--')  # Y-axis
plt.grid()
plt.legend()
plt.ylim(-3, 3)

# Show the plot
plt.show()
```
